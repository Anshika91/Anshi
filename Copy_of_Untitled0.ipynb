```python
# 5x1+2x2=9 ,3x1-7x2=5
# x1=(9-2x2)/5
#x2=(3x1-5)/7


x1=0
x2=0
x1g=x1
x2g=x2
error=1
count=0
while error>1.E-5:
  x1=(9-2*x2)/5
  x2=(3*x1-5)/7
  error=abs(x1-x1g)+abs(x2-x2g)
  plot(count,error,'k-o')
  x1g=x1
  x2g=x2
  count+=1
  print('x1:',round(x1,2),'x2:',round(x2,2))

 
```
